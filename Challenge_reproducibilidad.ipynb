```python
import tsplib95

# Cargar los archivos
xqf131 = tsplib95.load('xqf131.tsp')
pbl395 = tsplib95.load('pbl395.tsp')
xql662 = tsplib95.load('xql662.tsp')
dkg813 = tsplib95.load('dkg813.tsp')
xit1083 = tsplib95.load('xit1083.tsp')

# Coordenadas de los nodos
nodos131 = xqf131.node_coords
nodos395 = pbl395.node_coords
nodos662 = xql662.node_coords
nodos813 = dkg813.node_coords
nodos1083 = xit1083.node_coords
```

```python
import pandas as pd
df131 = pd.DataFrame.from_dict(nodos131, orient='index', columns=['x', 'y'])
df131.index.name = 'node'
df395 = pd.DataFrame.from_dict(nodos395, orient='index', columns=['x', 'y'])
df395.index.name = 'node'
df662 = pd.DataFrame.from_dict(nodos662, orient='index', columns=['x', 'y'])
df662.index.name = 'node'
df813 = pd.DataFrame.from_dict(nodos813, orient='index', columns=['x', 'y'])
df813.index.name = 'node'
df1083 = pd.DataFrame.from_dict(nodos1083, orient='index', columns=['x', 'y'])
df1083.index.name = 'node'
```

```python
print(f"Archivo cargado: {xqf131.name}")
print(f"Total de nodos: {len(df131)}")
df131.head()
```

```python
df131.describe()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 3, 1)
sns.scatterplot(x='x', y='y', data=df131, s=15, color='royalblue')
plt.title(f"Distribución de nodos: {xqf131.name}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.subplot(2, 3, 2)
sns.scatterplot(x='x', y='y', data=df395, s=15, color='seagreen')
plt.title(f"Distribución de nodos: {pbl395.name}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.subplot(2, 3, 3)
sns.scatterplot(x='x', y='y', data=df662, s=15, color='orange')
plt.title(f"Distribución de nodos: {xql662.name}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.subplot(2, 3, 4)
sns.scatterplot(x='x', y='y', data=df813, s=15, color='crimson')
plt.title(f"Distribución de nodos: {dkg813.name}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.subplot(2, 3, 5)
sns.scatterplot(x='x', y='y', data=df1083, s=15, color='purple')
plt.title(f"Distribución de nodos: {xit1083.name}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.show()
```
